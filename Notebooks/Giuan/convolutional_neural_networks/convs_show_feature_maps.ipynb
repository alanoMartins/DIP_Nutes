```python
model = cnn.ConvNet()
cnn.load_model(model, "./mnist_weights.pth")
```

```python
img = utils.get_image("./data/training/4/342.png")

# mudando o shape da imagem, Conv2d precisa que a imagem venha no formato
# batch x canais x altura x largura, mas a imagem com o formato altura x largura x canais 
image = torch.tensor([img], dtype=torch.float).permute(0, 3, 1, 2)

output = model(image)

element, index = torch.max(output[0], 0)
print(index)
```
