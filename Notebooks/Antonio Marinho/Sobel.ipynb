### importando bibliotecas


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('sudoku.png',0)
```

Exemplos de matrizes que ser√£o mutiplicadas pela imagem

$$

\begin{array}{ccc}
1 & 2 & 1  \\

\right]
$$

$$
GX = \left[
\begin{array}{ccc}

\end{array}
\right]
$$

```python
sobelx = cv2.Sobel(img,cv2.CV_64F,1,0,ksize=5)
sobely = cv2.Sobel(img,cv2.CV_64F,0,1,ksize=5)
```

### Exibindo imagem com o filtro Sobel  X

```python

plt.subplot(1,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(sobelx,cmap = 'gray')
plt.title('Sobel X'), plt.xticks([]), plt.yticks([])

plt.show()
```

### Exibindo imagem com o filtro Sobel  Y

```python
plt.subplot(1,2,1),plt.imshow(img,cmap = 'gray')
plt.title('Original'), plt.xticks([]), plt.yticks([])
plt.subplot(1,2,2),plt.imshow(sobely,cmap = 'gray')
plt.title('Sobel Y'), plt.xticks([]), plt.yticks([])

plt.show()
```
